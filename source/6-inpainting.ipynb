```python
import imageio
import numpy as np

def image_loads(paths):
    def load_(p):
        img = Image.open(p).convert('RGB')
        return np.array(img)

    results = [load_(e) for e in paths]

    return results

output_images = image_loads(outputs)

def generate_mp4(out_name, images, kwargs):
    writer = imageio.get_writer(out_name , **kwargs)
    for image in images:
        writer.append_data(image)
    writer.close()
```

```python

generate_mp4('../results.mp4', output_images, {'fps':24})
```
